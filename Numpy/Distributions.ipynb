### Random module

```python
import numpy as np

# generate a random integer from 0  to 100
x = np.random.randint(100)
print(x)
```

```python
# generate a random float

x= np.random.rand()
x
```

```python
# generate a random array - size parameter specify shape of the array

# 1D array containing 5 random integers from 0 to 100
x = np.random.randint(100, size=(5))
x
```

```python
# generate 2D array with 3 rows, each row containing 5 random integers from 0 - 100

x = np.random.randint(100, size=(3, 5))
x
```

```python
# Generate a random number from Array
# choice() method - allows to generate a random value on an array values
# choice() method - taken an array as parameter and randomly returns on of the values

# return one of the value in an array
x = np.random.choice([3, 5, 7, 9])
x
```

```python
# generate 2D array that consists of vlaues in array parameter (3,5,7, 9)

x = np.random.choice([3, 5, 7, 9], size=(3, 5))
x
```

##### Data Distribution
- A list of all possible values and how often each value occurs

##### Random Distribution
- A set of random numbers that follow a certain probability density function

##### Probabilty Density Function
- Describes a continuous probability i.e probability of all values in an array

```python
# 1D array with 100 values, where each value is 3, 5, 7,or 9
# probability for the value of 3 is 0.1
# probability for the value of 5 is 0.3
# probability for the value of 7 is 0.6 
# probability for the value of 9 is 0.0

x = np.random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))
x
```

```python
# 2D array with 3 rows, each containing 5 values
# probability for the value of 3 is 0.1
# probability for the value of 5 is 0.3
# probability for the value of 7 is 0.6 
# probability for the value of 9 is 0.0

x = np.random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))
x
```

##### Random Permutations
- arrangement of elements in an order

```python
# shuffle() - makes changes to the original array, 

arr = np.array([1, 2, 3, 4, 5])

np.random.shuffle(arr)
arr
```

```python
# permutation() randomly arranges the array elements
# returns re-arranged array and original array is unchanged
arr = np.array([1, 2, 3, 4, 5])

print(np.random.permutation(arr))
```

##### Normal Distribution - Gaussian Distribution
- Fits probability distribution of many events
- random.normal() - to get normal distribution
- loc - Mean where the peak of the bell exists
- scale -  standard deviation, how flat the graph distribution should be
- size - The shape of the returned array

```python
# Generate a random normal distribution of size 2x3
# with mean 1 and stddev 2
x = np.random.normal(loc = 1, scale=2, size=(2,3))
x
```

```python
# visualize the normal distribution
# The curve of a Normal Distribution is also known as the Bell Curve

sns.displot(np.random.normal(size=1000), kind="kde")
plt.show()
```

##### Binomial Distribution - Discrete Distribution
- describes outcomes of binary scenarios - toss of a coin
- n: number of trials
- p: probability of occurence of each trial
- size: shape of the returned array


```python
# 10 trials for a coin toss generate 10 data points

x = np.random.binomial(n=10, p=0.5, size=10)

x
```

```python
sns.displot(np.random.binomial(n=10, p=0.5, size=1000))
plt.show()
```

```python
# Normal distributoin is continous, binomial distribution is discrete

data = {
    "normal": np.random.normal(loc=50, scale=5, size=1000),
    "binomail": np.random.binomial(n=100, p=0.5, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

##### Poisson Distribution
- Discrete Distribution
- Estimates how many an event can happen in a specified time
- lam : rate or known number of occurences
- size: shape of the returned array

```python
# generate a random 1x10 distribution of occurence 2

x= np.random.poisson(lam = 2, size=10)
x
```

```python
# Visalize Poisson 

sns.displot(np.random.poisson(lam = 2, size=1000))

plt.show()
```

```python
# Normal distributoin is continous, poisson distribution is discrete

data = {
    "normal": np.random.normal(loc=50, scale=5, size=1000),
    "poisson": np.random.poisson(lam = 50, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```
