```python
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show
```

```python
# by default plot() draws a line from point to point
# Parameter 1 is an array of points on x axis - horizontal axis
# Parameter 2 is an array of points on y axis - vertical axis

x = np.array([1, 8])
y = np.array([3, 10])
plt.plot(x, y, 'o')
plt.show()
```

```python
# multiple points
# lines from (1, 3) to (2, 8) then to (6, 1) and to (8, 10)
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)
plt.show()
```

```python
# default x-points: 0, 1, 2, 3 depending on length of y
y = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y)
plt.show()
```

```python
# marker argument: emphasize each point with marker

y = np.array([3, 8, 1, 10])

plt.plot(y, marker = 'o')
plt.show()
```

```python
# marker argument: emphasize each point with marker

y = np.array([3, 8, 1, 10])

plt.plot(y, marker = '*')
plt.show()
```

```python
# fmt - parameter to specify the marker

# marker argument: emphasize each point with marker

y = np.array([3, 8, 1, 10])

plt.plot(y,'o-.r')
plt.show()
```

```python
# marker argument: emphasize each point with marker

y = np.array([3, 8, 1, 10])

# plt.plot(y,'*:b', ms=20, mec = 'r') # marker edge color
# plt.plot(y, 'o--g', ms = 20, mfc = 'r')     # marker face color
plt.plot(y, 'o:y', ms=20, mec='r', mfc='g')
plt.show()
```

```python
# keyword linestyle or ls, to change the style of plotted line

y = np.array([3, 8, 1, 10])

plt.plot(y, linestyle = 'dotted')
plt.show()
```

```python
# Multiple plots - with the subplot() function

# plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

# the figure has 1 row, 2 cols, and this plot is the first plot
plt.subplot(1, 2, 1)
plt.plot(x, y)
plt.title("SALES")

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

# the figure has 1 row, 2 cols, and this plot is the 2nd plot
plt.subplot(1, 2, 2)
plt.plot(x, y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

```python
# multiple plot on top of each other

# plot 1: 
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x, y)

# plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x, y)

plt.show()
```

```python
# bar() function to draw bar graphs

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width=0.1)        # default width is 0.8
plt.show()
```

```python
# bar() function to draw bar graphs

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# horizontal bars
plt.barh(x, y, color = "hotpink")
plt.show()
```

```python
# Histogram - shows frequency distribution
# Histogram show number of obseravations within each given interval

# hist() - uses an array of numbers to create an histogram

# generate an array of 250 values, where values concentrate around 170,
# and the standard deviation is 10

x = np.random.normal(170, 10, 250)

# the hist() function will read the array and produce the histogram
plt.hist(x)
plt.show()
```

```python
# PIE CHARTS
# each piece is called a wedge for each value in the array
# by default the first wedge starts fromx axis and moves coutner-clockwise
# size of each wedge is determined by comparing the value with other values
# using this formula -> x/sum(x)


y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels=mylabels, startangle=90, explode= myexplode, shadow=True, colors=mycolors)

plt.legend(title = "Four Fruits")
plt.show()
```
